```python
index = tagCnt.index
data = {}
for i in index:
    s = i.split(',')
    d = []
    for j in tags:
        if j in s:
            d.append(1)
        else:
            d.append(0)
    data[i] = d
data
```

```python
index = problems.index
df = problems.copy()
df = df[tags]
tagNames = []
for i in index:
    data = df.loc[i]
    tagName = ''
    for j in tags:
        if data[j] == 1:
            tagName = tagName+j+","
    tagName = tagName[:-1]
    tagNames.append(tagName)
df['tag'] = tagNames
df = df[['tag']]
df
            
```

```python
valueCnt = df['tag'].value_counts()
vdf = pd.DataFrame(valueCnt)
vdf.columns = ['cnt']
vdf.to_csv("tagCnt.csv")
```

```python
tags = problems.columns[:-1]
for t in tags:
    df = problems[problems[t]==1]
    df.to_csv("Tag/"+t+".csv")
```

```python
dic = {}
for i in tags:
    df = pd.read_csv("Tag/" + i + ".csv", index_col=0)
    data = []
    for t in tags:
        uni = df[t].unique()
        if 1 in uni:
            cnt = df[t].value_counts()[1]
            data.append(cnt)
        else:
            data.append(0)
    dic[i] = data
dic
```

```python
tdf = pd.DataFrame(dic, index = tags)
tdf.to_csv("tagDataFrame.csv")
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
#적절한 군집수 찾기
# inertia(군집 내 거리제곱합의 합) value(적정 군집수)
# df = problems.copy()
# df.drop('level', axis=1, inplace=True)
df = tagDF.copy()
ks = range(1,200)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(df)
    inertias.append(model.inertia_)

plt.figure(figsize=(40,40))

plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
model = KMeans(n_clusters=70)
model.fit(df)
centers = model.cluster_centers_
predict = model.predict(df)
print(centers)
print(predict)
```

```python
cdf = df.copy()
cdf['tag'] = predict
cdf
```
