```python
import pandas as pd
```

```python
problems = pd.read_csv('problems.csv', index_col = 0)
problems
```

```python
tagCluster = pd.read_csv('tagCluster.csv', index_col = 0)
tagCluster
```

```python
tagDF = pd.read_csv('tagDF.csv', index_col=0)
tagDF
```

```python
tagDF['tag'] = tagCluster['tag']
tagDF
```

```python
index = problems.index
index
```

```python
col = tagDF.columns[:-1]
col
```

```python
problemDF = problems[col]
problemDF
```

```python
tagNames = []
for i in index:
    data = problemDF.loc[i]
    tagName = ''
    for j in col:
        if data[j] == 1:
            tagName = tagName+j+","
    tagName = tagName[:-1]
    tagNames.append(tagName)
problemDF['tagName'] = tagNames
problemDF
```

```python
tagCluster = pd.read_csv('tagCluster.csv')
tagCluster.columns = ['tagName','tag']
tagCluster
```

```python
df = pd.merge(problemDF, tagCluster, how='left', on='tagName')
df
```

```python
df.index = index
df
```

```python
df['level'] = problems['level']
df
```

```python
levels = []
for i in index:
    level = df.loc[i]['level']
    if level<=5:
        levels.append('B')
    elif level<=10:
        levels.append('S')
    elif level<=15:
        levels.append('G')
    else:
        levels.append('P')
levels
    
```

```python
df['levelName'] = levels
df
```

```python
df[df['level'] == 11]
```

```python
df = df[['tag','levelName']]
```

```python
tags = []
for i in index:
    tag = str(df.loc[i]['tag']) + df.loc[i]['levelName']
    tags.append(tag)
tags
```

```python
category = pd.DataFrame(tags, columns = ['category'])
category['pId'] = index
category = category[['pId', 'category']]
category
```

```python
category.to_csv('category.csv')
```
