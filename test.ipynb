```python
levels = []
for i in range(len(cates)):
    levels.append(cates[i][-1]) 
levels
```

```python
level_count = {'B':0, 'S':0, 'G':0, 'P':0}
for i in levels:
    if i == 'B':
        level_count['B'] += 1
    elif i == 'S':
        level_count['S'] += 1
    elif i == 'G':
        level_count['G'] += 1
    elif i == 'P':
        level_count['P'] += 1
level_count
```

```python
cluster_names = {}
for j in range(70):
    tag_item = tag[tag['tag'] == j].index
    tag_item
    arr  = []
    for i in tag_item:
        items = i.split(',')
        for item in items:
            arr.append(item)
    arr
    tag_item_count = {}
    for i in arr:
        if i not in tag_item_count:
            tag_item_count[i] = 1
        else:
            tag_item_count[i] += 1
    tag_item_count
    sorted_tag_item_count = sorted(tag_item_count.items(), key = lambda x : x[1], reverse=True)
    cluster_name = ''
    for i in sorted_tag_item_count[:5]:
        cluster_name = cluster_name + i[0] + ','
    cluster_name = cluster_name[:-1]
    cluster_names[j] = cluster_name
cluster_names
```

```python
cluster = pd.DataFrame(cluster_names.values(), columns=['name'])
cluster
```
